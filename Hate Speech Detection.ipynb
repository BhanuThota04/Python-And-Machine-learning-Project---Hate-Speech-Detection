```python
dataset["labeled"]=dataset["class"].map({0:"Hate speech",1:"Offencive Speech",2:"No Hate or Offencive Speech"})
```

```python
data=dataset[["tweet","labeled"]]
```

```python
import re
import nltk
import string
nltk.download('stopwords')
#remove of stop words and steeming the words
```

```python
from nltk.corpus import stopwords
stopwords = set(stopwords.words("english"))
```

```python
stemmer=nltk.SnowballStemmer("english")
```

```python
def clean_data(text):
    txt=str(text).lower()
    text=re.sub('https?://\S+|www\.S+', '',text)
    text=re.sub('\[.*?\]', '', text)
    text=re.sub('<.*?>+', '', text)
    text=re.sub('[%s]' %re.escape(string.punctuation), '', text)
    text=re.sub('\n', '',text)
    text=re.sub('\w*\d\w*', '',text)
    text=[word for word in text.split(' ') if word not in stopwords]
    text=" ".join(text)
    text=[stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
```

```python
data["tweet"] = data["tweet"].apply(clean_data)
```

```python
data
```

```python
X=np.array(data["tweet"])
y=np.array(data["labeled"])
```

```python
X
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
cv=CountVectorizer()
X=cv.fit_transform(X)
```

```python
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
#Building ML Model
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
#confution matrix and accuracy
from sklearn.metrics import confusion_matrix
dt.fit(X_train,y_train)
y_pred=dt.predict(X_test)
cm=confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(cm, annot = True, fmt="d", cmap="YlGnBu")
plt.show()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
